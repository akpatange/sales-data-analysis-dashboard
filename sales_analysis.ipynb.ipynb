#    

# KPIs

## Total Sales
        Total sales represents the total revenue genrated from all orders within the selected time period.
Formula :- 
        Sum of the sales column.

```python
total_sales = df["sales"].sum()
print(f"Total Sales: ${total_sales:,.2f}")
```

 ## Interpretation
            This KPI shows overall business revenue performance. Higher total sales indicate strong market demand and business growth

#     

## Total Profit
        Total Profit represents the total earnings after deducting costs from sales.
Formula :- Sum of all Profit values

```python
total_profit = df["profit"].sum()
print(f"Total Profit : ${total_profit:,.2f}")
```

## Interpretation
        This KPI shows whether the company is actually making money. High sales with low Profit may indicate high costs or heavy discounting

#  

## Profit Margin
        Profit Margin measures how much profit is earned for each unit of sales
Formula :- (Total Profit / Total Sales) * 100

```python
profit_margin = (total_profit / total_sales) * 100
print(profit_margin)
```

## Interpretation 
        A higher profit margin indicates better cost control and pricing  strategy. 
        Low margin may signal operational inefficiencies or excessive discount

#  

## Total Order
        Total number of orders placed during the time period
Formula :- Count of unique order id

```python
total_order = df["order_id"].nunique()
print(total_order)
```

## Interpretation
        This KPI reflects customer demand and transaction volume. Growing order count often signals business expansion.

# 

## Average Order Value
        Average revenue genrated per order.
formula :- Total sales / Total orders

```python
average_order_value = (total_sales / total_order)
print(average_order_value)
```

## Interpretation
        Higher average order value suggests customers are buying more expensive or multiple products per order.

#  

# Advanced Kpi Insights

## Monthly Growth Rate(%)
        Percentage change in sales compared to the previous month.
Formula:- (current month sales - previous month sales) / previous month sales * 100

```python
df["month"] = df["order_date"].dt.to_period("M")
monthly_sales = df.groupby("month")["sales"].sum().reset_index()
print(monthly_sales)
```

```python
monthly_sales["Growth %"] = monthly_sales["sales"].pct_change()*100
monthly_sales
```

## Interpretation 
        Positive growth indicates expansion, while negative growth highlights potential issues requiring investigation.

# SALES PERFORMANCE ANALYSIS

## MONTHLY SALES TREND
            

```python
plt.plot(monthly_sales["month"].astype(str),monthly_sales["sales"])
plt.xticks(rotation=45)
plt.title("Monthly Sales Trend")
plt.show()
```

## 📌 Insight
        Sales show a consistent upward trend during Q4, with a peak in November, likely due to festive season demand.
        However, a noticeable decline is observed in February, indicating potential seasonal slowdown.
## 💡 Business Recommendation
            - Increase marketing efforts during low-performing months.
            - Leverage festive periods with targeted promotional campaigns.

## SALES BY REGION

```python
region_sales = df.groupby("region")["sales"].sum().reset_index()
```

```python
region_sales.sort_values("sales").plot(
    x="region", y="sales", kind="bar"
)
plt.title("Sales By Region")
plt.show()
```

## 📌 Insight
        The Central region generates the highest sales contribution, accounting for a significant share of total revenue.
        The Canada region underperforms compared to others.
## 💡 Business Recommendation
        - Investigate expansin opportunities in high-performing regions.
        - Analyze customer behavior and marketing effectiveness in low-performing regions.

#  

## PROFIT BY CATEGORY

#  

## TOP 10 PRODUCTS

```python
top_products = df.groupby("product_name")["sales"].sum().nlargest(10)
```

```python
top_products.plot(kind="bar")
plt.title("Top 10 Product By Sales")
plt.show()
```

## 📌 Insight
        The top 10 products contribute significantly to overall revenue, indicating strong customer demand for these items.
        A small group of products drives a large portion of total sales, suggesting a pareto effect (80/20 rule).
## 💡 Business Recommendation
        - Increase inventory for top-performing products.
        - Bundle top products with low-performing products.
        - Run targeted promotions on high-demand products.

            - Develop targeted promotions to grow Home Office sales.

# Correlation Heatmap

```python
sns.heatmap(df[["sales","profit","quantity","discount"]].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

## 📌 Insight
        The analysis indicates that while sales moderately contribute to profit, discounting has a noticeable negative impact on profitability
        without significantly boosting sales or quantity.
        This suggests that the current pricing and dicount strategy may be reducing margin without driving meaningful revenue growth.
## 💡 Business Recommendations
        - Re-evaluate discount policies to protect profit margins.
        - Focus on optimizing pricing strategy instead of aggressive discounting.
        - Promote high-margin products to improve overall profitability.
